# for Graph A

```python
import networkx as nx

# Read in graph data from text file
G = nx.read_edgelist('graph_A.txt', nodetype=int)

# Calculate degree centrality for each node
degree_centrality = nx.degree_centrality(G)

# Calculate betweenness centrality for each node
betweenness_centrality = nx.betweenness_centrality(G)

# Calculate closeness centrality for each node
closeness_centrality = nx.closeness_centrality(G)

# Calculate eigenvector centrality for each node
eigenvector_centrality = nx.eigenvector_centrality(G)

# Calculate page rank centrality for each node
page_rank_centrality = nx.pagerank(G)

# Print out the centrality measures for each node
print("node\tDegree Centrality\tBetweenness Centrality\tCloseness Centrality\tEigenvector Centrality\tPagerank Centrality")
for node in G.nodes():
    dc = degree_centrality[node]
    bc = betweenness_centrality[node]
    cc = closeness_centrality[node]
    ec = eigenvector_centrality[node]
    pc =  page_rank_centrality[node]
    print(f"{node}\t\t\t{dc:.3f}\t\t\t{bc:.3f}\t\t\t{cc:.3f}\t\t\t{ec:.3f}\t\t\t{pc:.3f}")

#save the centrality measures in a dataframe and node
import pandas as pd
df = pd.DataFrame({'degree_centrality':degree_centrality, 'betweenness_centrality':betweenness_centrality, 'closeness_centrality':closeness_centrality, 'eigenvector_centrality':eigenvector_centrality, 'page_rank_centrality':page_rank_centrality})

#rename the first column to node
df.index.name = 'node'

df.to_excel('centrality_measures_graph_A.xlsx')


df
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Load graph from file
G = nx.read_edgelist('graph_A.txt', nodetype=int)

# Find communities using asyn_lpa_communities method

communities = nx.algorithms.community.asyn_lpa_communities(G)


# Define color map for communities
color_map = {}
colors = ['red', 'blue', 'green', 'orange', 'purple', 'pink', 'brown', 'grey','aqua','cyan']
i = 0
for comm in communities:
    for node in comm:
        color_map[node] = colors[i]
    i += 1
    if i == len(colors):
        i = 0

# Draw graph and make distance between nodes 
plt.figure(figsize=(13,13))
nx.draw(G, node_color = [color_map.get(node) for node in G], with_labels=True, node_size=1000, font_size=15, font_color='black', width=0.3, alpha=0.7, pos=nx.spring_layout(G, k=0.5, iterations=30))
#make space between nodes
plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0, hspace=0)

# Show graph
plt.show()


```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Load graph from file
G = nx.read_edgelist('graph_A.txt', nodetype=int)

# Remove node 619 and its friends and friends of friends
remove_nodes = [619]
for friend in G.neighbors(619):
    remove_nodes.append(friend)
    for fof in G.neighbors(friend):
        remove_nodes.append(fof)
remove_nodes = set(remove_nodes)
G.remove_nodes_from(remove_nodes)

# Get the largest connected component
largest_cc = max(nx.connected_components(G), key=len)
G = G.subgraph(largest_cc)

# Check if the resulting graph is connected
is_connected = nx.is_connected(G)
if is_connected:
    print("The graph is connected.")
else:
    print("The graph is not connected.")

# Find communities using asyn_lpa_communities method
communities = nx.algorithms.community.asyn_lpa_communities(G)

# Define color map for communities
color_map = {}
colors = ['r', 'b', 'g', 'y', 'm', 'c', 'k']
i = 0
for comm in communities:
    for node in comm:
        color_map[node] = colors[i]
    i += 1
    if i == len(colors):
        i = 0

# Plot graph with communities colored differently show labels also increase the size of the graph

nx.draw(G, node_color = [color_map.get(node) for node in G], with_labels=True, node_size=1000, font_size=8, font_color='black', width=0.3, alpha=0.6, pos=nx.spring_layout(G, k=0.5, iterations=30))
#make space between nodes
plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0, hspace=0)

# Show graph
plt.show()

#print the nodes of the graph
print(G.nodes())
```

## Average degree, average clustering coefficient, average path length, 
## diameter, highest degree node, lowest degree node etc. of the graphs. (you 
## can make a table to present these measures.)

```python
import networkx as nx

# Load graph from file
G = nx.read_edgelist('graph_A.txt', nodetype=int)

# Calculate measures
avg_degree = sum(dict(G.degree()).values()) / len(G)
avg_clustering_coef = nx.average_clustering(G)
avg_path_length = nx.average_shortest_path_length(G)
diameter = nx.diameter(G)
highest_degree_node = max(dict(G.degree()), key=lambda x: dict(G.degree())[x])
lowest_degree_node = min(dict(G.degree()), key=lambda x: dict(G.degree())[x])

# Print measures
print('Average Degree: ', avg_degree)
print('Average Clustering Coefficient: ', avg_clustering_coef)
print('Average Path Length: ', avg_path_length)
print('Diameter: ', diameter)
print('node with highest degree: ', highest_degree_node)
print('node with lowest degree: ', lowest_degree_node)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Load the graph from a file
G = nx.read_edgelist('Graph_AB.txt')

# Degree distribution
degree_values = [val for node, val in G.degree()]
plt.hist(degree_values, bins=range(max(degree_values) + 2), align='left')
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.show()

# Clustering coefficient distribution
cc_values = list(nx.clustering(G).values())
plt.hist(cc_values, bins=10)
plt.xlabel("Clustering coefficient")
plt.ylabel("Frequency")
plt.show()

# Path length distribution
pl_values = nx.shortest_path_length(G)
pl_dict = dict(pl_values)
pl_values = [pl for node_pl in pl_dict.values() for pl in node_pl.values()]
plt.hist(pl_values, bins=max(pl_values))
plt.xlabel("Path length")
plt.ylabel("Frequency")
plt.show()
```

```python
G = nx.read_edgelist('graph_AB.txt', nodetype=int)

degree_sequence = sorted([d for n, d in G.degree()], reverse=True)
print("Degree sequence:", degree_sequence)

cc_sequence = sorted(nx.clustering(G).values(), reverse=True)
print("Clustering coefficient sequence:", cc_sequence)

pl_sequence = []
for u in G:
    for v in G:
        if u != v:
            pl_sequence.append(nx.shortest_path_length(G, u, v))
print("Path length sequence:", pl_sequence)
```
